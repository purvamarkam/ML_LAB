```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_samples = 1000
female_mean = 152
male_mean = 166
std_dev = 5

# Generate female and male height data
female_heights = np.random.normal(female_mean, std_dev, n_samples)
male_heights = np.random.normal(male_mean, std_dev, n_samples)

# Create DataFrame to label gender (F for female, M for male)
female_labels = ['F'] * n_samples
male_labels = ['M'] * n_samples

# Combine the heights and labels
heights = np.concatenate([female_heights, male_heights])
labels = np.concatenate([female_labels, male_labels])

# Create a DataFrame
data = pd.DataFrame({'Height': heights, 'Gender': labels})

# Visualize the distributions
plt.figure(figsize=(10, 6))
plt.hist(female_heights, bins=30, alpha=0.7, label='Female (Mean=152, SD=5)', color='blue')
plt.hist(male_heights, bins=30, alpha=0.7, label='Male (Mean=166, SD=5)', color='red')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Height Distribution for Males and Females')
plt.legend()
plt.grid(True)
plt.show()

# Preview the first few rows of the generated data
print(data.head())
```
